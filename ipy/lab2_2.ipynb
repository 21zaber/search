## Extract tokens rate from file

```python
fname = '/home/zaber/work/search/files/tokenrate10000'
data = []

with open(fname, 'r') as f:
    for i in f.readlines():
        t, r = i.split()
        data.append(int(r)) 
```

## Some imports

```python
%matplotlib inline

import math
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

## Zipf law

```python
x = [i for i in range(len(data))]
s = max(data)

fx = [i/s for i in data]
gx = [1/(i+1) for i in x]

plt.subplot(111)
plt.plot(gx, label='1/x')
plt.semilogy(x, fx, label='rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.show()
```

## Zipf–Mandelbrot law

```python
x = [i for i in range(len(data))]
#s = max(data)

p, q, b = 10**8, 5, -1.45

g = lambda x: p*(x+q)**b

fx = [i for i in data]
gx = [g(i) for i in x]

plt.subplot(111)
plt.plot(gx, label='g(x)')
plt.semilogy(x, fx, label='rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.show()
```
